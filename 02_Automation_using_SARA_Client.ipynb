<h1>Sentinel Australasia (Copernicus Hub/NCI)</h1>

http://www.copernicus.gov.au/data-access/


<h2>SARA Client access through AusCopHub_SearchSara.py</h2>

Visit the following to register:

https://copernicus.nci.org.au/sara.client/#/home

For more details see:

https://auscophub.readthedocs.io/en/latest/clientside.html

https://copernicus.nci.org.au/sara.client/#/help


Parameters are fully described at:

https://copernicus.nci.org.au/sara.server/1.0/api/collections/S2/describe.xml


Gives verbose XML description of the parameters available.

Example search command:

<b>auscophub_searchSara.py -q "box=148.0,-36.0,150.0,-35.0" --sentinel 2 -q "startDate=2018-08-20" -q "completionDate=2018-08-26" --curlscript curl-script.bash --jsonfeaturesfile feature-file.geojson</b>

Returns the datasets available bounded by the precified region between the dates specified.

Note: Need .netrc file in home directory containing:

<b>machine  copernicus.nci.org.au login <your_email_addres>  password <your_password></b>
    
Your email address and your password are those you chose when registering for your account.
    
(see https://everything.curl.dev/usingcurl/netrc for more info)
    
<br/>
<b>More complex searches</b>
    
Set up arguments and pass them to the search script.  
    
<b>'auscophub_searchSara.py',
'-q', 'box='+bbox_string,
'--sentinel', '3',
'-q','startDate='+start_date, 
'-q', 'completionDate='+completion_date,                          
'-q', 'productType=OL_1_EFR___',
'-q', 'instrument=OLCI', 
'--curlscript', "curl-script.bash", 
    '--jsonfeaturesfile', 'feature-file.geojson'</b>

Other useful arguments: 
    
    --resolution
    --processingLevel
    --cloudCover
    --waterCover
    --floodedCover
    
    --polygonfile shapefile (arbitrary polygons)
    --lon longitude           (circular searches)     
    --lat latitude 
    --radius radius_distance
    --geometry <wkt_polygon>
        e.g. POLYGON((92 29,110 29,110 6,115 6,115 11,119.1 11,119.1 22,129.3 22,129.3 -25.5,
                141.5 -25.5,141.5 -70,75 -70,75 -2.14,92 -2.14,92 6,92 29))
