<h1>Introduction to ASDAF - Day 2</h1>

Main data hub: http://www.asdaf.space/data/



<h2>National Map</h2>

http://asdaf.space/data/nationalmap/

Allows the user to:

- Search for and download data.
- Upload data.
- Add WMS/WCS/WMTS layers to desktop applications.
- View data in 2D or 3D.
- Compare datasets using split views.

<h2>Digital Earth Australia</h2>

http://asdaf.space/data/digital-earth-australia/

Hosts numerous satellite datasets aimed at mapping physical changes like erosion, vegetation change, water quality and availability, urban change, etc.

<h2>Sentinel Australasia Regional Access (SARA)</h2>

http://asdaf.space/data/copernicus-hub/

Provides free access to all Sentinel data for the South-East Asia and South Pacific region. 
- Can search for data using date range, satellite platform, sensor, etc.
- Can restrict search by drawing a region of interest.
- Has a Python API to facilitate automation.

<h2>Australian Ocean Data Network</h2>

http://asdaf.space/data/aodn/

Provides access to all available Australian marine and climate science data as well as access to Australia’s Integrated Marine Observing System (https://imos.org.au/)
