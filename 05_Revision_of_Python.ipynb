```python
# Variables

# Variables are dynamically typed

myint = 8
print(type(myint))

myfloat = 3.1415927
print(type(myfloat))

x = 8
print(type(x))

x = 3.1415927
print(type(x))
```

```python
# Conditions

# Comparison operators (==, !=, >, <, >=, <=)

a = 1
b = 2
condition = b > a

# Note the colon
if condition:
    #code for case where condition is true
    print("b > a")
else:
    #code for case where condition is true
    print("b <= a")

    
a = 10 
b = 5
condition = b > a

if condition:
    #code for case where condition is true
    print("b > a")
else:
    #code for case where condition is true
    print("b <= a")

    
# Can have multiple "else" conditions

# if cond1:
#     code for cond1
# elif cond2:
#     code for cond2
#
#       etc.
#
# else:
#     code for fallback case

    
    
```

```python
# Conditions

# Logical operators (not, and, or)

condition = True
print(not condition)

a = 1
b = 10
x = 3

print(x >= a and x <= b) # x is in domain [a,b]
```

```python
# Loops

print("Loop 1")
for i in range(10): # NOTE: Doesn't include last value
    print(i)
    
print("\nLoop 2")
for i in range(10, 20, 2):
    print(i)
```

```python
# Lists - creating and accessing elements

# Creating a list
suburbs = ["West Perth", "Cannington", "Applecross"]
print(suburbs)

# Accessing list elements
print(suburbs[0])  # First value in list
print(suburbs[1])  # Second value in the list
print(suburbs[-1]) # Last value in list
```

```python
# Create list
suburbs = ["West Perth", "Cannington", "Applecross"]
print("Before")
print(suburbs)

# Remove specified element from list
suburbs.remove("Applecross")
print("After")
print(suburbs)
```

```python
# Create list
suburbs = ["West Perth", "Cannington", "Applecross"]
print("Before")
print(suburbs)

# Empty the list
suburbs = []
print("After")
print(suburbs)
```

```python
# Create list
suburbs = ["West Perth", "Cannington", "Applecross"]

# Check item is in the list
if "Cannington" in suburbs:
    print("Cannington is in the list")
else:
    print("Cannington is not in the list")
    
if "Albany" in suburbs:
    print("Albany is in the list")
else:
    print("Albany is not in the list")
    
```

```python
import math

# List comprehensions - a succinct way of creating lists

odd = [ x for x in range(10) if x % 2 == 1]

print("\nOdd numbers")
print(odd)

even =  [ x for x in range(10) if x % 2 == 0]
print("\nEven numbers")
print(even)

squares = [x * x for x in range(10)]
print("\nSquares")
print(squares)


import os  # Modules can be imported at any time

# Note: we could get the list of files using os.listdir(".") for example
file_list = ["file1.txt", "file2.tif", "file3.jp2", "file4.doc", "file5.tif"]
tif_files = [ f for f in file_list if f.endswith(".tif")]
print("\nTIFF files")
print(tif_files)

other_files = [ f for f in file_list if not f.endswith(".tif")]
print("\nOther files")
print(other_files)
```

```python
# Functions

def squared(n):
    return n * n


def get_file_ext(file_name):    
    file_ext = file_name.split(".")[-1]
    return file_ext


print("squared(3)")
print(squared(3))

print("\nFile extension for test.txt")
print(get_file_ext("text.txt"))
```
